## Create Classes

```python
#Naive Bayes Classifier
class NaiveBayesClassifier:
    def __init__(self, feature_size):
        self.feature_size=feature_size
        self.mean=np.zeros(feature_size)
        self.var=np.zeros(feature_size)

    def fit(self, x, t):
        print("fitting... ")
        x_c1, x_c2 = self.split_classes(x,t)
        x_transposed = np.transpose(x_c1)
        for i in range(len(x_transposed)):
            self.mean[i] = np.mean(x_transposed[i])
            self.var[i] = np.var(x_transposed[i])
        print("Mean")
        print(self.mean)
        print("Variance")
        print(self.var)
        return

    def split_classes(self, x, t):
        mask1=(t==1)
        mask2=(t==-1)
        c1=x[mask1]
        c2=x[mask2]
        return c1, c2

    def predict(self, datapoint):
        probability=1;
        for i in range(len(datapoint)):
            gaussian_distribution = (1/ np.sqrt(2*np.pi*self.var[i])) * np.exp(-(datapoint[i] - self.mean[i])**2/(2 * self.var[i]))
            probability = probability * gaussian_distribution
        return probability
```

```python
#Gaussian Generative Model Classifier
class GMLE:
    def __init__(self,n_classes, feature_size):
        self.n_classes = n_classes
        self.feature_size=feature_size
        self.mk=np.zeros((n_classes, self.feature_size))
        self.invcov=np.zeros((self.feature_size,self.feature_size))
        
    def fit(self, x, ts):
        print("fitting... ")
        print(x.shape)
        cov=np.zeros((self.feature_size,self.feature_size))
        for i in range(self.n_classes):
            xi=self.split_classes(x,ts[i])
            mi=np.mean(xi, axis=0)
            self.mk[i]=mi
            covi = self.calc_Sk(xi,mi)
            cov+=(xi.shape[0]/(x.shape[0]))*covi
        self.invcov = np.linalg.pinv(cov)
        return
    
    def split_classes(self, x, t):
        mask1=(t==1)
        c1=x[mask1]
        return c1
        
    def calc_Sk(self, x1, m1):
        print('calculating cov...')
        nk=x1.shape[0]
        set1 = x1-m1
        term1=np.zeros((self.feature_size,self.feature_size))
        for a in set1:
            term1+=np.multiply((a),np.transpose(np.array([a])))
        return term1

    def calc_power(self, x, i):
        part = x-self.mk[i]
        term = np.matmul(np.matmul(np.array([part]), self.invcov), np.array([part]).T)
        return term

    def predict(self, x):
        preds=[]
        for i in range(self.n_classes):
            term = self.calc_power(x, i)
            p_cond_k = np.exp(-0.5*term[0][0])
            preds.append(p_cond_k)
        return np.argmax(preds)
```

## Create Helper Function

```python
#Prediction done using the method given at the bottom of slide 18 in Lecture 3.pdf
def predict(y_airplane, y_automobile, y_frog):
        if(y_airplane>=y_automobile and y_airplane>=y_frog):
            return np.array([1,0,0])
        if(y_automobile>y_airplane and y_automobile>=y_frog):
            return np.array([0,1,0])
        else:
            return np.array([0,0,1])
```

```python
def create_cm(y_actual, y_pred):
    cm=np.zeros((3,3))
    for i in range(len(y_pred)):
        real=int(y_actual[i])
        pred=int(y_pred[i])
        cm[pred][real]+=1
    return cm
```

```python
def read_train_rgb(name, size):
    #train_path = '/content/Data/train/'
    train_path = 'Data/train/'
    data = np.array([plt.imread(train_path+name+'/{}.jpg'.format(i)) for i in np.arange(0,5000)])
    if(size!=32):
        data = np.array([cv2.resize(data[i], (size,size), interpolation=cv2.INTER_AREA) for i in np.arange(0,5000)])
    data=data.reshape(-1,size*size*3)
    data=data/255
    return data
def read_test_rgb(name, size):
    #train_path = '/content/Data/train/'
    test_path = 'Data/test/'
    data = np.array([plt.imread(test_path+name+'/{}.jpg'.format(i)) for i in np.arange(0,1000)])
    if(size!=32):
        data = np.array([cv2.resize(data[i], (size,size), interpolation=cv2.INTER_AREA) for i in np.arange(0,1000)])
    data=data.reshape(-1,size*size*3)
    data=data/255
    return data
def create_labels(order):
    pos= np.ones(5000)
    neg =np.ones(5000)*-1
    if (order==0):
        labels = np.concatenate((pos,neg,neg))
    if order==1:
        labels = np.concatenate((neg,pos,neg))
    if order==2:
        labels = np.concatenate((neg,neg,pos))
    return labels
```

## Comment on the results ##

### 
Gaussian MLE performed better than NB as it relies on a continuous probability distribution which could account better for the relationship between colours as spectrum rather than independent values as is the case when using NB. With the exception of the 32x32 size, the smaller the image size, the lower the accuracy and this is due to having less features to classify based on. 32x32 gives a low accuracy which could possibly be due to noise which was reduced when downsizing.
